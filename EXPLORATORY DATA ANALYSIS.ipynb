```python
def model_logic(cols): #we are taking two columns
    
    Age=cols[0]
    
    Pclass=cols[1]
    
    
    if pd.isnull(Age):#where ever the age is null
        
        if Pclass == 1:
            return 39
        
        elif Pclass == 2:
            return 29
        
        else:
            return 25
    else:
        return Age

```

```python
train['Age']=train[['Age','Pclass']].apply(model_logic,axis=1)


#apply all the value of train data set of column age and pclass into the function
```
